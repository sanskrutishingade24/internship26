```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([10,23,29,34,39,45,50])
y = np.array([14,21,31,43,48,55,58])
z = np.array([56,50,44,37,26,20,16])
```

```python
#line chart
```

```python
plt.plot(x,y);
```

```python
plt.xlabel("x values")
plt.ylabel("y values")
plt.title("line chart: x vs y")
plt.plot(x,y,marker = 'o', mfc = 'red',ls = '--');
plt.grid();
```

```python
plt.xlabel("y values")
plt.ylabel("z values")
plt.title("line chart: y vs z")
plt.plot(y,z,marker = '>', mfc = 'blue',ls = '--');
plt.grid();
```

```python
plt.xlabel("x values")
plt.ylabel("z values")
plt.title("line chart: x vs z")
plt.plot(x,z,marker = '>', mfc = 'orange');
plt.grid();
```

```python
plt.xlabel("x values")
plt.ylabel("y values")
plt.title("line chart: y vs z")
plt.plot(x,y,marker = '>', mfc = 'blue',ls = '--');
plt.plot(x,z,marker = 'o',mfc = 'black',ls = '--');
plt.grid();
```

```python
plt.figure(figsize=(16,9))

plt.subplot(2,2,1)
plt.xlabel("x values")
plt.ylabel("y values")
plt.title("line chart: x vs y")
plt.plot(x,y,marker = 'o', mfc = 'red',ls = '--');
plt.grid();

plt.subplot(2,2,2)
plt.xlabel("x values")
plt.ylabel("z values")
plt.title("line chart: x vs z")
plt.plot(x,z,marker = 'o', mfc = 'red',ls = '--');
plt.grid();
```
