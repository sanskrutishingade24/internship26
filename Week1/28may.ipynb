```python
num = int(input("Enter a number: "))

factorial = 1
i = 1

if num < 0:
    print("Factorial does not exist for negative numbers")
else:
    while i <= num:
        factorial *= i
        i += 1

    print(f"Factorial of {num} is {factorial}")
```

```python
# Python program to find sum of digits of a number

num = int(input("Enter a number: "))

sum_digits = 0

while num > 0:
    digit = num % 10
    sum_digits += digit
    num = num // 10

print("Sum of digits =", sum_digits)
```
